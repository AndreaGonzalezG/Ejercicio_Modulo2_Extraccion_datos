```python
def sacar_tabla_medallero(url):

    #PASOS INICIALES
    #  url de la página de la vamos a sacar datos
    url_medallero = "https://es.wikipedia.org/wiki/Anexo:Medallero_de_los_Juegos_Olímpicos_de_Tokio_2020"
    # Request a la página 
    res_medallero = requests.get(url_medallero)
    # vemos si todo ha ido bien
    print("La respuesta de la petición es:", res_medallero.status_code)
    # creamos el objeto BeautifulSoup 
    sopa_medallero = BeautifulSoup(res_medallero.content, 'html.parser')
    
    #BUSQUEDA DE INFO EN LA WEB
    # Con el metodo ".find_all()" busco table
    tablas = sopa_medallero.find_all("table")

    #La  tabla que me interesa
    tabla_medallero = tablas[0]

    # Encabezados/ seleccionaremos la etiqueta "th" usando el método "th"
    lista_encabezados = tabla_medallero.find_all("th")
    encabezados_medallero = [columna.text.replace('\n', '') for columna in lista_encabezados]

     # Ajuste de encabezados
    encabezados_medallero[2]= 'oro'
    encabezados_medallero[3]= 'plata'
    encabezados_medallero[4]= 'bronce'
    

    # Filas de la  tabla de medalleros, usando el método .find_all()
    filas_medallero = tabla_medallero.find_all("tr")
    # Creamos una lista para almacenar los resultados obtenidos.
    resultados_medallero = []

    # Iniciamos un bucle 'for' para iterar a través de la lista pero comenzamos desde el segundo elemento de la lista.
    for fila in filas_medallero[1:]:
        # Para cada 'fila', extraemos el texto, lo dividimos en una lista usando '\n' como separador y eliminamos el primer y último elemento de la lista.
        fila_texto = fila.text
        elementos_fila = fila_texto.split("\n")[1:-1]

        # Añadimos la lista de elementos a la lista 'resultados_medallero'.
        resultados_medallero.append(elementos_fila)


    # Limpiamos y procesamos los datos en 'resultados_medallero'

    for indice_lista, lista in enumerate(resultados_medallero):
        for indice_elemento, elemento in enumerate(lista):
            # Verifica si el elemento es una cadena antes de intentar limpiarlo
            if isinstance(elemento, str):
                # Elimina los espacios no deseados (\xa0) y quita espacios en blanco alrededor
                elemento_limpio = elemento.replace("\xa0", "").strip()
            else:
                # Si el elemento no es una cadena, lo dejamos tal como está
                elemento_limpio = elemento
            
            # Reasignamos el elemento procesado a la lista original
            resultados_medallero[indice_lista][indice_elemento] = elemento_limpio

    return encabezados_medallero,resultados_medallero
```

```python
encabezados_medallero_final, datos_medallero_final = sacar_tabla_medallero(url_medallero)
```

```python
# convertimos la lista de listas a DataFrame usando el método 'pd.DataFrame()'. 
df_medallero = pd.DataFrame(datos_medallero_final)
df_medallero
```
